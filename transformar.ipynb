```python
import pandas as pd
```

```python
df = pd.read_csv('datos/ipc_mes.csv', sep=';', encoding='iso-8859-1', index_col=False)
df
```

```python
df.drop(['Grupos ECOICOP', 'Tipo de dato'], axis = 1, inplace = True)
df 
```

```python
dic = {}

for index, row in df.iterrows():
    anyo = row['Periodo'][:4]
    mes = int(row['Periodo'][-2:])
    comunidad = row['Comunidades y Ciudades Autónomas']
    
    if comunidad not in dic:
        dic[comunidad] = {f'{a}T{t}': 0 for a in range(2017, 2020) for t in range(1,5)}
    
    trim = ((mes-1) // 3) + 1
    dic[comunidad][f'{anyo}T{trim}']+=float(row['Total'].replace(',', '.'))/3
    
print(dic)
```

```python
df_col = {
    'Comunidades y Ciudades Autónomas': [],
    'Periodo': [],
    'Total': []
}
df = pd.DataFrame(df_col)

for comunidad, x in dic.items():
    for periodo, total in x.items():
        datos = {
            'Comunidades y Ciudades Autónomas': comunidad,
            'Periodo': periodo,
            'Total': total
        }
        df = pd.concat([df, pd.DataFrame([datos])], ignore_index=True)
```

```python
df
```

Vamos a añadir el resto de dataframes

```python
df_activos = pd.read_csv('datos/activos_trim.csv', sep=';', encoding='iso-8859-1')
df_activos.drop(['Sexo', 'Edad'], axis = 1, inplace = True)
df_activos
```

```python
df_vivienda = pd.read_csv('datos/indicesCCAA_trim.csv', sep=';', encoding='iso-8859-1')
df_vivienda.drop(['Índices y tasas', 'General, vivienda nueva y de segunda mano', 'Total Nacional'], axis = 1, inplace = True)
df_vivienda
```

```python
merged_df1 = pd.merge(df, df_activos, on=['Comunidades y Ciudades Autónomas', 'Periodo'])
merged_df = pd.merge(merged_df1, df_vivienda, on=['Comunidades y Ciudades Autónomas', 'Periodo'])
merged_df
```

Vamos a renombrar las columnas

```python
new_column_names = {
    'Total_x': 'IPC', 
    'Total_y': 'Personas_activas',
    'Total': 'Precio_vivienda',
    'Comunidades y Ciudades Autónomas': 'CCAA'
}
merged_df.rename(columns=new_column_names, inplace=True)
```

Podemos ver que las dos columnas no procesadas no tienen el formato numérico adecuado, vamos a cambiarlo.

```python
# Replace the commas with periods and remove thousands separators
merged_df['Personas_activas'] = merged_df['Personas_activas'].str.replace(',', '.').str.replace('.', '', n=1)
merged_df['Precio_vivienda'] = merged_df['Precio_vivienda'].str.replace(',', '.').str.replace('.', '.', n=1)

merged_df
```

```python
def delete_first_three_chars(s):
    return s[3:]

# Apply the function to all elements in the 'Column1' column
merged_df['CCAA'] = merged_df['CCAA'].apply(delete_first_three_chars)

merged_df
```
